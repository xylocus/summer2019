```python
"""
We also have some of these data broken down by demographic (race, age, income, etc).
The tricky part is that the same variable can mean different things in different year,
thus we need to build a quite complicated query.
"""
# for these queries, there are too many variables, you will first need to get metadata
meta = load_data.load_outcome_metadata()
meta.head()
```

```python
meta['groupby_type'].unique()
```

```python
meta['topic'].unique()
```

```python
# example query
meta[
    (meta['groupby_type']=='race') &
    (meta['topic']=='income') &
    (meta['type']=='estimate') &
    (meta['desc']=='median income (dollars)')
]['groupby'].unique()
```

```python
asian_income = meta[
    (
        (meta['groupby']=='one race-- - asian') |
        (meta['groupby']=='households - one race-- - asian')
    ) &
    (meta['topic']=='income') &
    (meta['type']=='estimate') &
    (meta['desc']=='median income (dollars)')
]
asian_income
```

```python
# first load the "all" dataset
all_msa = load_data.load_outcome_data(geo_level = 'msa', dataset = 'all')
```

```python
# with these variable_name + topic + year combination, you can pandas-fu your way to the data you need
# but this function is also provided for your convenience
load_data.get_data_from_metadata(asian_income, all_msa, 'msa')
```
